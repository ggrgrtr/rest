ГРАД СПУСК


```python
w=q.tensor([[5.,10.],[1.,2.]],requires_grad=True)
alpha=0.001


for i in range(500):
  # critical: calculate the function inside the loop
  F=q.log(q.log(w+7)).prod() # нам нужно обновление функции, т.к. тензор обновляется -> тензор привязывается к функции
  F.backward()
  w.data=w-alpha*w.grad # обновление тензора
  w.grad.zero_()

w

```

```python
w = q.tensor([8., 8.],requires_grad=True)

optimizer = q.optim.SGD([w], lr=0.001)

def parabola(x):
  return 10*(x**2).sum()

def grad_step(F,tensor1):
  F_res = F(tensor1)
  F_res.backward()
  # tensor1.data -= 0.001*tensor1.grad   --- equals
  # tensor1.grad.zero_()
  optimizer.step()
  optimizer.zero_grad()

for i in range(500):
  grad_step(parabola, w)

w
```
